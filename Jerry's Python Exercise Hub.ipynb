```python
z = ["a","b","c"]
dummy_vector = np.random.randn(10,3)
print(z)
print("dummy vector before normalization : \n", dummy_vector)

def normalizeRows(x):
    """ Row normalization function

    Implement a function that normalizes each row of a matrix to have
    unit length.
    """

    """Jerry 
    the purpose of this method is to normalize values for a give matrix X, normalize meanning to return matrix value between 0..1
    x: input matrix
    N = x.shape[0], meaning to return the number of elements for dimension 0
    X**2, meaning square of X, after X**2 the matrix value all become square of X, then sum(,axis=1) meaning to sum the value of each column, that is sum the row.
    sqrt, meaning square root
    using of X**2 is to offset the sign effect of positive and negative
    "1e-30", because normalization use divide function, therefore this is to ensure non-zero in the devide operation.
    """
    N = x.shape[0]
    print("N:",N)
    print("x**2: \n",x**2)
    print("np.sum(x**2, axis=1): \n",np.sum(x**2, axis=1)) # axis=1 meaning to add the value on axis1 - column, so the return should have shape(10;)
    print("np.sqrt(np.sum(x**2, axis=1)): \n",np.sqrt(np.sum(x**2, axis=1)))
    print("np.sqrt(np.sum(x**2, axis=1)).reshape((N,1)): \n",np.sqrt(np.sum(x**2, axis=1)).reshape((N,1)))  #. reshape similar to transpose.before reshape it is one dimensional vector, reshap it to two dimensinoal vector
    print("np.sqrt(np.sum(x**2, axis=1)).reshape((N,1)) + 1e-30: \n",np.sqrt(np.sum(x**2, axis=1)).reshape((N,1)) + 1e-30)
    print("x before return: \n",x)
    x /= np.sqrt(np.sum(x**2, axis=1)).reshape((N,1)) + 1e-30

    return x

dummy_vector_option1 = normalizeRows(dummy_vector)
dummy_vector_option2 = dummy_vector / np.linalg.norm(dummy_vector, axis=1, keepdims=True)



print("dummy vector after normalization Option 1 : \n", dummy_vector_option1)
print("dummy vector after normalization Option 2 : \n", dummy_vector_option1)


# Ensure they are effectively identical with a tolerance 1e-8
assert np.allclose(dummy_vector_option1, dummy_vector_option2, atol=1e-8), "The matrices are not effectively identical"

print 
# Ensure they are identical
assert np.array_equal(dummy_vector_option1, dummy_vector_option2), "The matrices are not identical"

```

### <font color='blue'> Exercise 3: build sigmoid test function </font>
1) you are not familiar with sigmoid activation function, so you want to build a test function and use some concrete numbers
to understand the behavior that probability approach to 1 if x>0, and probablitliy approach to 0 if x<0. 
2) The sigmoid function maps any real-valued number to the range (0, 1).

Prompt:write a python example for sigmoid function, then help adjust it to give an input vector with 500 random  numbers betweeen -100 and 100,  sort it from smallest to largest, and then loop the vector and calcute sigmoid result for each element, and finally plot the result in a graph.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate a vector of 500 random numbers between -100 and 100
x_values = np.random.uniform(-100, 100, 500)

# Sort the values in ascending order
x_values_sorted = np.sort(x_values)

# Calculate the sigmoid for each element
sigmoid_values = sigmoid(x_values_sorted)

# Plot the results
plt.plot(x_values_sorted, sigmoid_values, label="Sigmoid(x)")
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.title('Sigmoid Function for Sorted Random Values')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate a vector of 500 random numbers between -5 and 5
x_values = np.random.uniform(-5, 5, 50)

# Sort the values in ascending order
x_values_sorted = np.sort(x_values)

# Calculate the sigmoid for each element
sigmoid_values = sigmoid(x_values_sorted)

# Plot the results
plt.plot(x_values_sorted, sigmoid_values, label="Sigmoid(x)")
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.title('Sigmoid Function for Sorted Random Values')
plt.grid(True)
plt.legend()
plt.show()

# Create a DataFrame to display the table of values
data = {'x': x_values_sorted, 'Sigmoid(x)': sigmoid_values}
df = pd.DataFrame(data)

# Print the table of x values and corresponding sigmoid values
print(df)
```
